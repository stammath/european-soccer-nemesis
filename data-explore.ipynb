### all available teams and their league

```python
teams_per_league = pd.read_sql("""SELECT Team.team_long_name as home_team_name,
                                         min( Country.name) as country_name_min,
                                        min(League.name) AS  league_name_min
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team on Match.home_team_api_id = Team.team_api_id
                                group by 1
                                order by 2
                                """, conn)
teams_per_league
```

### number of teams per league

```python
teams_per_league = pd.read_sql("""
select league_name, count(distinct home_team_name) as num_teams
from 
(SELECT Team.team_long_name as home_team_name,
                                         min( Country.name) as country_name,
                                        min(League.name) AS  league_name
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team on Match.home_team_api_id = Team.team_api_id
                                group by 1) 
                                group by 1 
                                order by 2 desc
                                """, conn)
teams_per_league
```

## list of all matches

```python
detailed_matches = pd.read_sql("""SELECT Match.id, 
                                        Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season, 
                                        stage, 
                                        date,
                                        HT.team_long_name AS  home_team,
                                        AT.team_long_name AS away_team,
                                        home_team_goal, 
                                        away_team_goal                                        
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                ORDER by date
                                LIMIT 10;""", conn)
detailed_matches
```

## number of matches per league

```python
matches_per_league = pd.read_sql("""select league_name, season, count(match_id) as num_matches
                                        from
                                    (   SELECT Match.id as match_id, 
                                        Match.season as season,
                                        Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season, 
                                        stage, 
                                        date                
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                ORDER by date) group by 1,2
                                order by 3 desc
                                ;""", conn)
matches_per_league
```
